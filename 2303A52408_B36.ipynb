Prediction of Online Shoppers Purchase Intentions.

 1.Name the product most liked and purchased

2. Identify the special days when most products are purchased

3. List the top 5 special days of purchase and type of visitors.

4. Identify the months when most purchases happened.

5. What the max and min revenue generated by the online shoppers

```python
import pandas as pd
df = pd.read_csv('/content/online+shoppers+purchasing+intention+dataset.zip')
print(df.columns)
most_purchased_product = df['ProductRelated'].value_counts().idxmax()
print(f"Product most liked and purchased: {most_purchased_product}")
```

```python
import pandas as pd
special_days_purchase = df[df["Revenue"]].groupby("SpecialDay").size()
print("\nPurchases on special days:")
print(special_days_purchase)
```

```python
import pandas as pd
top_special_days = (
    df[df["Revenue"]]
    .groupby(["SpecialDay", "VisitorType"])
    .size()
    .sort_values(ascending=False)
    .head(5)
)
print("\nTop 5 special days and visitor types:")
print(top_special_days)
```

```python
import pandas as pd
monthly_purchases = df[df["Revenue"]].groupby("Month").size().sort_values(ascending=False)  # Changed 'data' to 'df'
print("\nMonthly purchases:")
print(monthly_purchases)
```

```python
import pandas as pd
revenue_by_purchase = df[df["Revenue"]]["PageValues"]
max_revenue = revenue_by_purchase.max()
min_revenue = revenue_by_purchase.min()
print(f"\nMaximum revenue: {max_revenue}")
print(f"Minimum revenue: {min_revenue}")
```
